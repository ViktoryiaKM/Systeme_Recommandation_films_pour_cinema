# 🎬 Selection des Blockbusters avec Dataset **"tmdb_full"**

##  Objectif
Identifier les films considérés comme **blockbusters** à partir du dataset `tmdb_full`, selon des critères définis.
##  Critères de sélection

- **Revenu** (`revenue`) > seuil (à préciser)
- **Budget** élevé
- **Popularité** (selon score TMDB)
- **Note moyenne** éventuellement

👉 Ces critères permettent de filtrer les films ayant eu un **impact commercial important**.

## 3️⃣ Définition des critères de blockbuster

**!!! Pour info, La notation des films :Échelle de 0 à 10 (SensCritique, IMDb)**  
0 à 2 : Très mauvais (rejet massif, échec critique et public)  
2 à 4 : Mauvais (mal reçu par la critique et le public)  
4 à 6 : Moyen (réception mitigée, apprécié par certains, critiqué par d'autres)  
6 à 8 : Bon (généralement bien reçu, apprécié par la majorité)  
8 à 10 : Excellent (acclamation générale, grand succès critique et public)  

### Création de règles pour définir un **blockbuster**:
 - budget > 30 mln
 - recette > 30 mln
 - bonne notation: "vote_average" > 7,5
 - on selectionne les films ayant au moins 4000 "vote_count"

### Application des filtres:

```python
# Filtres sur le Revenue et/ou Budget:
df_blockbusters = df_tmdb_full[(df_tmdb_full['budget'] >= 30000000)|(df_tmdb_full['revenue'] >= 30000000)].sort_values(by='vote_average', ascending=False)
df_blockbusters.shape
```

```python
# avec des filtre sur les votes:
df_blockbusters1 = df_blockbusters[(df_blockbusters['vote_average'] >= 7.5) & (df_blockbusters['vote_count'] >= 3000)].sort_values(by='revenue', ascending=False)
df_blockbusters1.shape
```
