## 3Ô∏è‚É£ D√©finition des crit√®res de blockbuster

**!!! Pour info, La notation des films :√âchelle de 0 √† 10 (SensCritique, IMDb)**  
0 √† 2 : Tr√®s mauvais (rejet massif, √©chec critique et public)  
2 √† 4 : Mauvais (mal re√ßu par la critique et le public)  
4 √† 6 : Moyen (r√©ception mitig√©e, appr√©ci√© par certains, critiqu√© par d'autres)  
6 √† 8 : Bon (g√©n√©ralement bien re√ßu, appr√©ci√© par la majorit√©)  
8 √† 10 : Excellent (acclamation g√©n√©rale, grand succ√®s critique et public)  

### Cr√©ation de r√®gles pour d√©finir un **blockbuster**:
 - budget > 30 mln
 - recette > 30 mln
 - bonne notation: "vote_average" > 7,5
 - on selectionne les films ayant au moins 4000 "vote_count"

### Application des filtres:

```python
# Filtres sur le Revenue et/ou Budget:
df_blockbusters = df_tmdb_full[(df_tmdb_full['budget'] >= 30000000)|(df_tmdb_full['revenue'] >= 30000000)].sort_values(by='vote_average', ascending=False)
df_blockbusters.shape
```

```python
# avec des filtre sur les votes:
df_blockbusters1 = df_blockbusters[(df_blockbusters['vote_average'] >= 7.5) & (df_blockbusters['vote_count'] >= 3000)].sort_values(by='revenue', ascending=False)
df_blockbusters1.shape
```
