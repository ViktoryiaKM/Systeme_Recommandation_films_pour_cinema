### **1.1 Chargement d‚Äôun DataFrame "merge_global" complet pr√©alablement nettoy√©**

```python
import pandas as pd
import numpy as np
```

```python
# on charge notre dataframe globale:
merge_global = pd.read_csv("merge_global.csv", sep = ",")
merge_global
```

### **1.2 Application de filtres pour ne conserver que les films correspondant aux cat√©gories "famille", "enfants"**

**Travail effectu√©:**

- Suppression de la colonnes 'Unnamed: 0'
- Selection des filmes dont le genre contient "Comedy","Adventure",'Family', 'Animation', 'Documentary'
- Suppression des filmes dont le genre contient "Horror"
- Selection des filmes dont l'ann√©e de sortie > 1980
- Selection des filmes dont 'averageRating' > 6
- Selection des filmes dont 'numVotes' > 300

```python
merge_global['genres'].value_counts()
```

```python
merge_global.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# On selectionne les filmes dont le genre contient "Comedy","Adventure",'Family', 'Animation', 'Documentary'
df_enfants_famille = merge_global[(merge_global['genres'].str.contains('Comedy'))|(merge_global['genres'].str.contains('Family'))|(merge_global['genres'].str.contains('Adventure'))|(merge_global['genres'].str.contains('Animation'))|(merge_global['genres'].str.contains('Documentary'))]
```

```python
# On supprime les filmes dont le genre contient "Horror","Crime"
df_enfants_famille = df_enfants_famille[~df_enfants_famille['genres'].str.contains('Horror')]
df_enfants_famille = df_enfants_famille[~df_enfants_famille['genres'].str.contains('Crime')]
```

```python
# On selectionne les filmes dont 'averageRating' > 6
df_enfants_famille = df_enfants_famille[(df_enfants_famille['averageRating'] > 6)]
```

```python
# On selectionne les filmes dont 'averageRating' > 6
df_enfants_famille = df_enfants_famille[(df_enfants_famille['numVotes'] > 300)]
```

```python
# On selectionne les filmes dontl'ann√©e de sortie > 1980
df_enfants_famille = df_enfants_famille[(df_enfants_famille['startYear'] > 1980)]
df_enfants_famille
```

```python
# on v√©rifie s'il n'y a pas d'autre genres √† retirer:
# regrouper les genres par cat√©gorie, les compter et les classer du plus grand au plus petit : en affichant uniquement les 50 premi√®res cat√©gories
df_enfants_famille.groupby(by = 'genres')['genres'].count().sort_values(ascending = False)[:20]
```

```python
# on cherche les doublons (s'il y en a): => pas de doublons
duplicated_rows= df_enfants_famille.duplicated('tconst')
duplicated_rows.value_counts()
```

```python
df_enfants_famille.drop_duplicates(subset=['tconst'], inplace=True)
```

```python
df_enfants_famille_for_ML = df_enfants_famille.sort_values(by='averageRating', ascending=False)
df_enfants_famille_for_ML
```

## 2Ô∏è‚É£ Cr√©ation de variables pour le ML

```python
df_enfants_famille_for_ML = pd.read_csv('df_enfants_famille_for_ML.csv')
df_enfants_famille_for_ML.head(3)
```

```python
df_enfants_famille_for_ML.drop(columns = 'Unnamed: 0',axis=1,inplace=True)
```

### **2.1 Encodage des genres avec la m√©thode **get_dummies****

Pour rendre la colonne `"genres"` exploitable dans un mod√®le de Machine Learning, on la transforme en **variables binaires** (aussi appel√©es *dummies*) => Transformation de variables cat√©gorielles en variables binaires

####  **Objectif** est de cr√©er une colonne par genre, avec :
- `1` si le film appartient √† ce genre
- `0` sinon

```python
df_genres = df_enfants_famille_for_ML["genres"].str.get_dummies(sep=',')
df_genres
```

### **2.2 Fusion du DataFrame d'origine avec les colonnes dummies**

Apr√®s avoir transform√© la colonne `"genres"` en plusieurs colonnes binaires (`df_genres`), on les **r√©int√®gre** dans le DataFrame principal.

```python
# on concataine sur les colonnes le df_genres avec le df d'origne sans la colonne "genres"
df_dummies_enfants = pd.concat([df_enfants_famille_for_ML.drop(columns="genres"), df_genres], axis=1)
df_dummies_enfants
```

### **2.3 Encodage de la colonne "primaryName" (noms d‚Äôacteurs principaux)  avec la m√©thode **get_dummies****
**Objectif** est de cr√©er une colonne par "primaryName", avec :  
- 1 si le film appartient √† ce primaryName  
- 0 sinon


```python
# on applique Methode Dummies + tranformation en boolean
df_primaryName = df_dummies_enfants["primaryName"].str.get_dummies(sep=',')
df_primaryName
```

### **2.4 Fusion du DataFrame principal avec les colonnes dummies pour un Dataset final pour Machine Learning**

Apr√®s avoir transform√© la colonne `"primaryName"` en plusieurs colonnes binaires (`df_gprimaryName`), on les **r√©int√®gre** dans le DataFrame principal.

```python
# on concat sur les colonnes le df_genres avec le df d'origne sans la colonne "genres"
df_dummies_enfants_for_ML_def = pd.concat([df_dummies_enfants.drop(columns="primaryName"), df_primaryName], axis=1)
df_dummies_enfants_for_ML_def
```

```python
# il faut penser √† importer la df_dummies_enfants_for_ML_def.csv  .select_dtypes(include = ["int64", "float64"])

import pandas as pd
from sklearn.neighbors import NearestNeighbors
df_dummies_enfants_for_ML_def = pd.read_csv('df_dummies_enfants_for_ML_def.csv')
# df = df_dummies_enfants_for_ML_def.csv
df = df_dummies_enfants_for_ML_def

var_target = 'French_title'


X = df[['Action.1', 'Adventure.1', 'Animation.1', 'Biography.1', 'Comedy.1',
             'Documentary.1', 'Drama.1', 'Family.1', 'Fantasy.1', 'History.1',
             'Music.1', 'Musical.1', 'Mystery.1', 'News.1', 'Romance.1', 'Sci-Fi.1',
             'Sport.1', 'Thriller.1', 'War.1', 'Western.1','averageRating','runtimeMinutes','startYear']]
y = df[var_target]


modelNN = NearestNeighbors(n_neighbors=5, algorithm='auto')
modelNN.fit(X)

titre = "Ast√©rix chez les Bretons"
index_titre = df[df['French_title'] == titre].index[0]

distances, indices = modelNN.kneighbors([X.iloc[index_titre]])

print(f"Films similaires √† '{titre}':")
for idx in indices[0]:
    print(df[var_target].iloc[idx])
```
